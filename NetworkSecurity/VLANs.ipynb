# Evaluation Questions

a) WHY CAN'T YOU SEE THE PACKET SENT FROM HOST1 AT HOST3 TCPDUMP OUTPUT ?

Hosts cannot send frames outside their VLAN's. Since the VLAN of host1 and host2 is 150 and the VLAN of host3 and host4 is 250 , host 1 and host 3 are in different VLAN'S. Hence frame from host1 cannot reach host3.



b) WHY CAN YOU SEE THE 2 PACKETS SENT FROM HOST1 AT HOST5 ?

VLAN value for host5 is NONE and host5 is the default gateway. Both frames sent from host1 to any destination and to host2 is received at host5 because each time a frame is sent, it reaches its destination through default gateway. Also port4 is a trunk port connected to host5 and it carries traffic from all VLANs. Hence host5 gets frame from VLAN 150 and VLAN 250. 



c) WHAT DO YOU EXPECT WILL HAPPEN TO COMMUNICATION BETWEEN THE HOSTS THAT WERE IN THAT VLAN?

We are altering the value of the vlan of bridge1:1 to 250 which allows host2 to receive frames only from host3 or host4. Since VLAN of host1 is 150 the frame sent cannot reach host2 whereas it reaches host5 because it is default gatway.
