### Evaluation Question: What is the difference between an hostname and a fully qualified domain name (FQDN)? What is an NXdomain?

Hostnames are human readable nicknames that correspond to the address of a device connected to a network and is used to distinguish one device from another on a specific network or over the Internet.

A fully qualified domain name (FQDN) is the complete domain name for a specific computer, or host, on the Internet. The FQDN consists of two parts: the hostname and the domain name. For example, an FQDN for a hypothetical mail server might be mymail.somecollege.edu

Hostname: mymail
Domain: somecollege.edu
FQDN: mymail.somecollege.edu

The NXDOMAIN is a DNS message type received by the client when a request to resolve a domain is sent to the DNS and cannot be resolved to an IP address. An NXDOMAIN error message means that the domain does not exist. NXDOMAIN can also take place due to the network or DNS server problem.
